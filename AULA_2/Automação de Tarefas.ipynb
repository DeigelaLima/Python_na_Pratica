```python
# Para não ficar chamando 'yfinance', atribuo um apelido 'yf'
import yfinance as yf

# Carregando os Dados Históricos da 'Petrobras(PETR4.SA)' diretamente do Yahoo
# yf.Ticker("PETR4.SA").history("5d") # dados de cinco dia(5d)
# yf.Ticker("PETR4.SA").history("1y") # dados de um ano(1y)

# Guardando/Atribuindo na variável 'dados' as tabelas de dados que está sendo carregadas do Yahoo Finaças
# dados = yf.Ticker("PETR4.SA").history("6mo") # dados de seis meses(6mo)

ticker = input("Digite o código da Ação: ")
dados = yf.Ticker(ticker).history("6mo")
# Carregado toda a tabela
#dados

# Acessando as colunas da tabela
fechamento = dados.Close
#fechamento

# Cria um gráfico com o valor de fechamento da Ação
fechamento.plot()

# Obs.: no eixo y é o valor em real da Ação
# e no eixo x é o período de tempo
```

- Cotação máxima
- Cotação mínima
- Cotação atual

```python
# Cotação máxima: ponto/valor máximo de dentro da coluna Fechamento da tabela
fechamento.max()
```

```python
# Acessa o penúltimo índice
fechamento[-2]
```

```python
# O comando 'round' define quantas casas decimais o Valor Final vai possuir,
# passando por parâmentro a quantidade de casas decimais desejada.
maxima = round(fechamento.max(), 2)
minima = round(fechamento.min(), 2)
atual = round(fechamento[-1], 2)

print(maxima)
print(minima)
print(atual)
```

# Passp 3: enviar automaticamente o email com o resultado da análise

- Processo de enviar um e-mail passo a passo:
    - abrir uma nova aba no navegador (CTRL + T)
    - digitar o endereço do gmail (www.gmail.com) e digitar ENTER
    - clicar em Escrever
    - digitar o endereço de e-mail do destinatário
    - mudar para o campo Assunto (clicar no campo ou digitar TAB)
    - digitar o Assunto do e-mail
    - mudar para campo principal do e-mail (clicar no campo ou digitar TAB)
    - escrever a mensagem
    - clicar em Enviar

```python
# Automatiza o teclado e o mouse do PC
!pip install pyautogui
```

```python
# Copia texto para a área de transferência em Python
!pip install pyperclip
# Obs.: se tiver utilizando o Anaconda essa biblioteca já vem inclusa.
```

```python
import pyautogui
import pyperclip
```

```python
# dar uma pausa de 2 segundos entre os passos
# configurar uma pausa entre as ações do pyautogui
pyautogui.PAUSE = 2

# Abrir uma nova aba (CTRL + T)
# 'hotkey' é um atalho
pyautogui.hotkey("ctrl", "t")

# digitar www.gmail.com e dar um 'ENTER'
# pyperclip automatiza o 'CTRL + C' em texto (AUTOMATIZA ATALHOS)
# Copiou para memória (CTRL + C) 
# copiar o endereço do gmail para o clipboard
pyperclip.copy("www.gmail.com")

# colar(CTRL + V) o endereço do gmail e dar um ENTER
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

# clicando no botão Escrever
pyautogui.click(x=142, y=201)

# digitar o destinatário (TAB)
pyperclip.copy("deigelalima@hotmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# digitar o assunto (TAB)
pyperclip.copy("Análises diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# PRETENDO ENSINAR ESSE PROJETO EM ALGUM MINICURSO,
# POR ISSO ESTOU DEIXANDO ESSE EXEMPLO DE FORMA ESTÁTICA.
    # preencher o corpo do e-mail
    # mensagem = """

    # Prezado Gestor,

    # Seguem, conforme solicitado, as análises dos últimos seis meses da Ação XPTO:

    # Cotação máxima: R$ 27.18
    # Cotação mínima: R$ 19.11
    # Cotação atual: R$ 27.18

    # Qualquer dúvida, fico à disposição!

    # Att.
    # Deigela Lima

    # """

# preencher o corpo do e-mail
# EXEMPLO DE FORMA DINÂMICA
#esse 'f' indica que o texto será formatado dinamicamente
mensagem = f""" 

Prezado Gestor,

Seguem, conforme solicitado, as análises dos últimos seis meses da Ação {ticker}:

Cotação máxima: R$ {round(maxima,2)}
Cotação mínima: R$ {round(minima,2)}
Cotação atual: R$ {round(atual,2)}

Qualquer dúvida, fico à disposição!

Att.
Deigela Lima

"""

# copia o texto que está na variável mensagem
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

# Clicando no botão Eviar
pyautogui.click(x=1302, y=1045)

# fechar a aba do gmail
pyautogui.hotkey("ctrl", "f4")

# Imprimir mensagem de enviado com sucesso
print("E-mail enviado com sucesso!")
```
